```python
#Driver function for the graph visualization:

def draw_graph(G):
  nx.draw_networkx(G, pos = nx.kamada_kawai_layout(G), arrows = True if nx.is_directed(G) else None, node_size = 500, node_color = "yellow", font_size = 20)
```

```python
# Check set of nodes in the graph.
# Inputs:
# G: the networkx graph.

def checkNodes(G):
  print("Set of vertices: {0}".format(list(G.nodes())))
```

```python
checkNodes(random_graph)
```

```python
# Check set of edges in the graph.

def checkEdges(G):
  print("Set of Edges: {0}".format(list(G.edges())))
```

```python
checkEdges(random_graph)
```

```python
# Check degrees for the given nodes.

def checkDegrees(G, nodes_to_check_degree):
  for node in nodes_to_check_degree:
    if node in G.nodes():
      print("degree of {0}:{1}".format(node, G.degree(node)))
    else:
      print("Node {0} is not in Graph".format(node))
```

```python
checkDegrees(random_graph, [0, 1, 2, 10])
```

```python
# Check degree sequence for the graph:

def checkDegreeSequence(G):
  print("Degree sequence: {0}".format(sorted([G.degree(n) for n in G.nodes])))
```

```python
checkDegreeSequence(random_graph)
```

```python
# Combine all the basic graph properties in a single function:
# Properties:
# a. Vertices
# b. Edges
# c. Degree
# d. Degree-Sequence
def checkParameters(G, nodes_to_check_degree = None):
   print("Set of vertices: {0}".format(list(G.nodes())))
   print("Set of Edges: {0}".format(list(G.edges())))
   if nodes_to_check_degree is not None:
    for node in nodes_to_check_degree:
      if node in G.nodes():
        print("degree of {0}:{1}".format(node, G.degree(node)))
      else:
        print("Node {0} is not in Graph".format(node))
   print("Degree sequence: {0}".format(sorted([G.degree(n) for n in G.nodes])))
```

```python
checkParameters(random_graph, nodes_to_check_degree=[0, 1, 2, 10])
```

```python
walk_graph = nx.Graph()
walk_graph.add_edges_from([("d", "e"),("e", "f"), ("d", "a"), ("a", "e"), ("a", "b"), ("b", "e"), ("b", "f"), ("d", "c"), ("b", "c"), ("c", "f")])
```

```python
def checkWalkAndLength(G, possible_walk):
  is_walk = True
  walk_length = -1
  for i in range(len(possible_walk)- 1):
    if not G.has_edge(possible_walk[i], possible_walk[i+1]):
      is_walk = False
      break
  if is_walk:
    walk_length = len(possible_walk) - 1
  return is_walk, walk_length
```

```python
walk = 'a,b,a,b,a,b,a,b,b'
is_walk, walk_length = checkWalkAndLength(walk_graph, walk.split(','))
if is_walk:
  print("{0} is a walk of length{1}".format(walk, walk_length))
else:
  print("{0} is not a walk".format(walk))
```

```python
def checkPathAndLength(G, possible_path):
  is_path = True
  path_length = -1
  is_walk, walk_length = checkWalkAndLength(G, possible_path)
  if not is_walk:
    is_path = False
  else:
    path_edges = set()
    for i in range(len(possible_path)-1):
      possible_edge = (possible_path[i], possible_path[i+1])
      possible_edge1 = (possible_edge[1], possible_edge[0])
      if possible_edge in path_edges or possible_edge1 in path_edges:
        is_path = False
        break
      else:
        path_edges.add(possible_edge)
  if is_path:
    path_length = len(possible_path) - 1
  return is_path, path_length
```

```python
def checkFreemanFormula(G):
  n = len(G.nodes)
  max_degree = checkDegrees(G, G.nodes)
  for i in G.nodes:
    Cd += max_degree - checkDegrees(G, i)
  den = (n-1)*(n-2)
  return Cd/den
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a Barabási–Albert graph with 1000 nodes and m=5
n = 1000
m = 5
ba_graph = nx.barabasi_albert_graph(n, m)

# Get the degree of each node
degrees = [ba_graph.degree(node) for node in ba_graph.nodes()]

# Plot the degree distribution
plt.hist(degrees, bins=30, color='skyblue', edgecolor='black')
plt.title("Degree Distribution of Barabási–Albert Graph")
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.show()
```
